```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,f1_score
```

```python
digits=load_digits()
```

```python
dir(digits)
```

```python
digits.data[0]
```

```python
plt.gray()
for i in range(10):
    plt.matshow(digits.images[i])
```

```python
digits.target_names
```

```python
X=digits['data']
y=digits['target']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=115)
```

```python
logregM=LogisticRegression(multi_class='ovr')
```

```python
logregM.fit(X_train,y_train)
```

```python
predictions=logregM.predict(X_test)
predictions
```

```python
accuracy_score(y_test,predictions)
```

```python
precision_score(y_test,predictions,average='micro')
```

```python
recall_score(y_test,predictions,average='micro')
```

```python
f1_score(y_test,predictions,average='micro')
```

```python
logregM.score(X_test,y_test)
```

```python
plt.matshow(digits.images[1190])
```

```python
digits.target[1190]
```

```python
logregM.predict([digits.data[1190]])
```

```python
confusionMat=confusion_matrix(y_test,predictions)
confusionMat
```

```python
rows=len(confusionMat)
cols=len(confusionMat[0])
a=0
for i in range(rows):
    for j in range(cols):
        a=a+confusionMat[i][j]
a
```

```python
b=0
for i in range(rows):
    for j in range(cols):
        if i==j:
            b=b+confusionMat[i][j]
b
        
```

```python
accuracy_con_matrix=b/a
accuracy_con_matrix
```

```python
##Visualize
plt.figure(figsize=(12,8))
sns.heatmap(confusionMat,annot=True,cmap='viridis')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion_Matrix')
```
