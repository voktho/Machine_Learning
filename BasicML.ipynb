# MatplotLib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.arange(0,13)
y=np.arange(14,27)
```

```python
plt.scatter(x,y,c='b')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('2D Graph')
plt.savefig('2D Graph.jpg')
```

```python
plt.subplot(2,2,1)
a=x*x
plt.plot(x,a,'r*--',linewidth=3,markersize=13)
plt.subplot(2,2,2)
b=x^3
plt.plot(x,b,'g+-',linewidth=3,markersize=12)
plt.subplot(2,2,3)
c=x+y
plt.plot(x,c,'bo--',linewidth=3,markersize=5)
plt.subplot(2,2,4)
d=x^4
plt.plot(x,d,'y*-',linewidth=4,markersize=12)
```

```python
#Sinewave/cos
np.pi
x=np.arange(0,3*np.pi,0.01)
y=np.sin(x)
plt.plot(x,y,'r')
y1=np.cos(x)
plt.plot(x,y1,'g')
plt.legend(['Sine','Cosine'])
```

```python
##BarPlot
x1=[2,4,6]
yy=[3,9,4]
x2=[3,5,9]
yy2=[10,2,21]
plt.bar(x1,yy,color='g')
plt.bar(x2,yy2)
plt.title('Barplot')
```

```python
##histogram
p=np.array([2,3,1,4,2,5,6,7,5,14,7,13,14,2,3,1,9,12])
plt.hist(p)
plt.title('Histogram')
```

```python
##Boxplot
bxarr=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(bxarr,vert=True,patch_artist=True)
```

```python
#Pie Chart
labels='CSP','Electronics','Power'
sizes=[17,20,10]
explode=(0.1,0.05,0.02)
colors=['gold','Yellowgreen','lightcoral']
plt.pie(sizes,explode=explode,labels=labels,colors=colors,
         autopct='%1.1f%%',shadow=True)
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
dfle=pd.read_csv('carprices.csv')
```

```python
dfle['Car Model']=le.fit_transform(dfle['Car Model'])
dfle
```

```python
#loop....>>>if we have to encode more than once 
dflel=pd.read_csv('carprices.csv')
for col in dflel.columns:
    if dflel[col].dtype ==object:
        dflel[col]=LabelEncoder().fit_transform(dflel[col])
        
dflel
```
